# Functions

```python
# pythagoras_calculator = function name
# a,b = input parameters

def py_calculator(a,b):
    c = (a**2 + b**2)**0.5
    cc = round(c,2)
    return cc
```

```python
print (py_calculator(10,6))
```

```python
# explore how to filter AEIOU
def vocal_filter(text):
    filtered_str = ""
    for letter in text:
        if letter.lower() not in "aeiou":
            filtered_str += letter
    return filtered_str
```

```python
vocal_filter('Hello Good Morning everyone my name is wei yuan')
```

```python
no_aeiou = input("What is your name?: ")
vocal_filter(no_aeiou)
```
