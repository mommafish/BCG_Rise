# Warm up
- Create mean_calc function that takes 1 input of list, and return the mean
- Create stdev_calc function that takes 1 input of list and return the sample statdard deviation of sample

```python
# Create mean_calc function that takes 1 input of list, and return the mean
def mean_calc(e):    
    answer = sum(e)/(len(e))
    return answer
```

```python
excel_list = [1,3,5,7,9,12]
mean_calc(excel_list)
```

```python
# Create stdev_calc function that takes 1 input of list and return the sample statdard deviation of sample
# Sample Stdev
def stdev_calc(e):

    x_value = []
    for i in e:
        x_value.append((i - mean_calc(e))**2)
    
    return ((sum(x_value))/(len(e)-1))**0.5
```

```python
stdev_calc(excel_list)
```

```python
#1 open file from HDB
hdb = open('resale-flat-prices-based-on-registration-date-from-jan-2015-to-dec-2016.csv','r')

#2 readline(n) read first line as heading
headings = hdb.readline()

#3 readlines() ead all remaining lines (TAKE NOTE OF THE S at the end)
lines = hdb.readlines()

#4 prepare two empty list as placeholders, resale_price_list and town_list
town_list = []
resale_price = []

#5 use loop to append the values from the data into the list variables created before
for i in lines:
    sg = i.strip().split(',')
    town_list.append(sg[1])
    resale_price.append(float(sg[-1]))

#6 store both list into a dictonaray where the keys are 'resale_price and 'town' for each variable, respectively
my_dict = {'town':town_list, 'resale_price':resale_price}

#7 create a simple_stat function that takes two inputs: a town name and mode ('mean' or 'stdev' default is 'mean')
#8 the output of the function is the mean or stdev of the town
def simple_stat(town, mode='mean'):
    
    my_price = []
    for e in range(len(lines)):
        if my_dict['town'][e].lower() == town.lower():
            my_price.append( my_dict['resale_price'][e] )
            
    if mode.lower() == 'mean':
        return round(mean_calc(my_price),2)
    elif mode.lower() == 'stdev':
        return round(stdev_calc(my_price),2)
    else:
        return 'mode is invalid'


hdb.close()
```

## Question 1 define mean and stdev calculator

```python
# Stdev calculator using list comprehension
def stdev_calc(L):
    stdev = (sum( [ (X-mean_calc(L))**2 for X in L ]) / (len(L)-1) ) ** 0.5
    return stdev
```

```python
# Mean Calculator
def mean_calc(L):
    return sum(L)/len(L)
```

```python
# Stdev Calculator
def stdev_calc(L):
    mean = mean_calc(L)
    sum_of_square = 0
    
    for X in L:
        sum_of_square = sum_of_square + (X-mean)**2
    
    variance = sum_of_square/(len(L)-1)
    stdev = variance**0.5
    
    return stdev
```
