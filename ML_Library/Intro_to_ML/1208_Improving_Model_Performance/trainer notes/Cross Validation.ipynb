#### Types of Crossvalidation technique
1. Holdout Validation Approach
2. K-Fold Cross Validation
3. Stratified Cross Validation
4. Leave One Out Cross Validation
5. Repeated Random Train Test Split - ShuffleSplit

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(cancer.data,columns=cancer.feature_names)
data['target'] = cancer.target
data.head()
```

```python
x = data.drop(['target'],axis=1)
y = data['target']
```

```python
x.describe()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
pd.DataFrame(x).describe()
```

#### HoldOut Validation Approach

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.30,random_state=50)
model = LogisticRegression()
model.fit(xtrain,ytrain)
result = model.score(xtest,ytest)
print(result)
```

#### K- Fold Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np
model = LogisticRegression()
k_fold_validation = KFold(10)
results = cross_val_score(model,x,y,cv=k_fold_validation)
print(results)
print(np.mean(results))
```

#### Stratified Cross Validation

```python
data['target'].value_counts()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
skfold = StratifiedKFold(n_splits=10)
model = LogisticRegression()
results = cross_val_score(model,x,y,cv=skfold)
print(results)
print(np.mean(results))
```

#### Leave One Out Cross Validation

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
leave_vaildation = LeaveOneOut()
model = LogisticRegression()
results = cross_val_score(model,x,y,cv=leave_vaildation)
```

```python
print(results)
print(np.mean(results))
```

#### Repeated Train test - ShuffleSplit - Combination of train_test_split and K-Fold

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
shuffle_split = ShuffleSplit(n_splits=10,test_size=0.30)
model = LogisticRegression()
results = cross_val_score(model,x,y,cv=shuffle_split)
print(results)
print(np.mean(results))
```
