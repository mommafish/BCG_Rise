```python
# Split the data into train and test

from sklearn.model_selection import train_test_split

## independent variable = all, other fields
## dependent variable = survived = target = predictive variables

x_train, x_test, y_train, y_test = train_test_split(dfnew[['age','fare']],   # predictive
                                                    dfnew['survived'],       # Target
                                                    test_size = 0.3)  
```

```python
#3 wihtout deleting data we can fille with mean / median 
# missing values handling and fill with mean  
#df.horsepower.fillna(df.horsepower.mean(),inplace  = True  )
x_train['age_mean'] = x_train['age'].fillna(x_train.age.mean())
x_train['age_median'] = x_train['age'].fillna(x_train.age.median())
x_train.head()
```
