###  	2. Solve below using pandas dataframes
* Convert the Mileage into float number
* Find the count of cars whose Type=’Sedan’ & Category= ‘Sedan 4D’.
* Find the maximum mileage in each category within Type = 'Sedan'
* Find the average mileage of cars in each 'Type' whose price is greater than 15000.
* Find all details(Price,Mileage,Make,Model,Category,Type) of the cars  in the ‘Category’ = ‘CX Sedan’ whose price is greater than the highest pricing car in Category=’Sedan 4D’

Dataset: __cars.xls__

Read the data

```python
# import pandas as pd and read the data using read_excel
df = pd.read_excel('./Data/cars.xls')

# print top rows using head to inspect data
df.head()
```

Convert the Mileage into float number

```python
# check data types using dtypes
df.dtypes
```

```python
# convert 'Mileage column into float by using astype method and assigning it to the same column
df['Mileage'] = df['Mileage'].astype('float')

# print the data types again to see if it has changed
df.dtypes
```

Count of cars whose Type = 'Sedan' & Category = 'Sedan 4D'

```python
### Count of cars whose Type = 'Sedan' & Category = 'Sedan 4D'
# Create a list of conditions separately, it is a pandas series of True and False based on condition
find_index = (df['Type'] == 'Sedan') & (df['Category'] == 'Sedan 4D')

# create a new dataframe using the conditions
df2 = df[find_index]

# print the row count of the new dataframe
print("{} of cars have Type Sedan and Sedan 4D category".format(df2.shape[0]))
```

Find the maximum mileage in each category within Type = 'Sedan'

```python
# subset dataframe using the condition == for 'Type' column
# use groupby to group by the column 'Category'
# select the column 'Mileage' and find the maximum
df[df['Type']=='Sedan'].groupby('Category')['Mileage'].max()
```

Find the average mileage of cars in each 'Type' whose price is greater than 15000.

```python
# use filter to pick up rows having 'Price' greater than 15000
# group the dataframe by 'Type'
# select 'Mileage' and find the average value
df[df['Price']>15000].groupby('Type')['Mileage'].mean()
```

Find all details(Price,Mileage,Make,Model,Category,Type) of the cars  in the ‘Category’ = ‘Sedan 4D’ whose price is greater than the highest pricing car in Category=’CX Sedan 4D’

```python
# find the highest pricing of car in 'CX Sedan 4D' category
# use loc function to apply condition of Category == 'CX Sedan 4D' and select 'Price' column
# use max function to aggregate
highest_pricing_car =  df.loc[df['Category'] == 'CX Sedan 4D','Price'].max()

# filter for Category 'Sedan 4D' and also filter by price greater than highest price calculated earlier
g = df.loc[df['Category'] == 'Sedan 4D', : ]

# filter for price greater than highest price calculated earlier
res = g[g['Price']>highest_pricing_car] 
res
```

###  3. Solve below using pandas dataframes:
* What was the maximum temperature in New York in the month of January?
* On which day it rained?
* What is the average wind speed?

Dataset: __weather.csv__


```python
# Read the input file weather.csv
weather = pd.read_csv('./Data/weather.csv')

# convert 'EST' into datetime object assign it to a new column 'EST_Date'
weather['EST_Date']= pd.to_datetime(weather['EST'])

# Calculate the month of the EST_Date
weather['Month'] = weather['EST_Date'].dt.month

# Answer of I
# group by 'Month' calculated earlier, select 'Temperature' and find out maximum value
print(weather.groupby('Month')['Temperature'].max())
print("---"*30)

# Answer of II
# use loc and apply filter in row position to take only rows having Events as 'Rain'
# pick up 'EST_Date' column
# take only the unique values to get all dates only once
print(weather.loc[(weather['Events'] == 'Rain'), 'EST_Date'].unique())

print("---"*30)
# Answer of III
# print the mean of 'WindSpeedMPH'
# select the column and use mean function to get the average value
print(weather['WindSpeedMPH'].mean())
```

###  	4. Solve below using pandas dataframes:
* Select all the records (using dataset(airport.xlsx)) whose ‘iso_region ‘ = ‘US-CA’ & ‘type’ = ‘seaplane_base’
* Select ident, name, municipality (using dataset(airport.xlsx)) from all the records whose ‘iso_region ‘ = ‘US-CA’ & ‘type’ = ‘large_airport’
* Select all the records (using dataset(airport-frequencies.xlsx)) whose ‘airport_ident’ = ‘KLAX’ by sorting the records descendingly by ‘Type’

Datasets: __airports.csv__ and __airport-frequiencies.csv__

```python
import pandas as pd

airports = pd.read_csv('./Data/airports.csv')
airport_freq = pd.read_csv('./Data/airport-frequiencies.csv')

# Answer of I
print(airports.loc[(airports['iso_region'] == 'US-CA') & (airports['type'] == 'seaplane_base'),:].head())

print("-----"*30)

# Answer of II
print(airports.loc[(airports['iso_region'] == 'US-CA') & (airports['type'] == 'large_airport'),:].head())

print("-----"*30)

# Answer of III
print(airport_freq.loc[(airport_freq['airport_ident'] == 'KLAX'),:].sort_values(by = 'type', ascending = False).head())
```

###  	5. Manipulate Date & Time data 
* First convert the type of date to TimeStamp object.
* Change index to date column
* Retrieve Jan 2017 stock prices
* Average price of Apple stock in Jan 2017(Close field)
* Get me the prices of last week of Jan 2017

Dataset: __HistoricalQuotes_Apple_TimeSeries_Datetimeindex.csv__

```python
# Read the input csv file
df = pd.read_csv('./Data/HistoricalQuotes_Apple_TimeSeries_Datetimeindex.csv')
```

```python
# convert 'Date' column into datetime format
df['Date'] = pd.to_datetime(df['Date'])

# convert Date as index using set_index
df.set_index('Date', inplace=True)
```

```python
# Retrieve Jan 2017 stock prices
# use strftime function to compare index with '2017-Jan'
print(df[df.index.strftime("%Y-%b") == "2017-Jan"])
print("----"*20)
# use dataset created earlier, select column 'Close' and measure average
print(df.loc[df.index.strftime("%Y-%b") == "2017-Jan", 'Close'].mean())
print("----"*20)
# From the dataframe earlier in answer1, apply further filter of day greater than 24
# do it using strftime (%d) to indicate days and set datatype int to compare with 24
# select the column 'Close'
print(df.loc[(df.index.strftime("%Y-%b") == "2017-Jan") & (df.index.strftime("%d").astype(int)>24),'Close'])
```

#### Using the Titanic data, create histogram and kernel density estimate plots of the age distributions of survivors and victims.

```python
titanic = pd.read_excel('./Data/titanic.xls')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic.columns
```

```python
# create a title for the plot usint plt.title
# filter dataset for survived==1, select age and then use hist function to plot histogram
# do the same for survivied==0
# provide legend for the two plots
plt.title('KDE plot for Titanic Age distributions')
titanic.loc[titanic['survived']==1,'age'].hist()
titanic.loc[titanic['survived']==0,'age'].hist()
plt.legend(['Survived', 'Not Survived'], loc = 'best')
```

```python
# build kde plot to make it look neater
plt.title('KDE plot for Titanic Age distributions')
titanic.loc[titanic['survived']==1,'age'].plot.kde()
titanic.loc[titanic['survived']==0,'age'].plot.kde()
plt.legend(['Survived', 'Not Survived'], loc = 'best')
```
